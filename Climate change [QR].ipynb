## Temperature and CO2 emission

```python
#note: interpolated vs trend
co2data= pd.read_csv("https://raw.githubusercontent.com/quynh-anh-bang/You-live-and-learn/main/1_CO2-data.csv")
 #co2data["trend"] has the unit of month while global is in yr
```

```python
co2data
```

```python
global_58_17 = global_80_22[(global_80_22["Year"]>=1958)&(global_80_22["Year"]<=2017)]
global_59_17 = global_80_22[(global_80_22["Year"]>=1959)&(global_80_22["Year"]<=2017)]
global_58_16 = global_80_22[(global_80_22["Year"]>=1958)&(global_80_22["Year"]<=2016)]
global_concat_1 = pd.concat([global_59_17["Jan"],global_59_17["Feb"],global_58_17["Mar"],global_58_17["Apr"],global_58_17["May"],global_58_17["Jun"],global_58_17["Jul"],global_58_16["Aug"],global_58_16["Sep"],global_58_16["Oct"],global_58_16["Nov"],global_58_16["Dec"]])
```

```python
co2_1= co2data[co2data['month']==1]
co2_2= co2data[co2data['month']==2]
co2_3= co2data[co2data['month']==3]
co2_4= co2data[co2data['month']==4]
co2_5= co2data[co2data['month']==5]
co2_6= co2data[co2data['month']==6]
co2_7= co2data[co2data['month']==7]
co2_8= co2data[co2data['month']==8]
co2_9= co2data[co2data['month']==9]
co2_10= co2data[co2data['month']==10]
co2_11= co2data[co2data['month']==11]
co2_12= co2data[co2data['month']==12]
co2_concat = pd.concat([co2_1,co2_2,co2_3,co2_4,co2_5,co2_6,co2_7,co2_8,co2_9,co2_10,co2_11,co2_12])
```

```python
plt.scatter(co2_concat['interpolated'],global_concat_1)
plt.xlabel("CO2 emission (kilotons)")
plt.ylabel("Global temperature anomalies")

a,b = np.polyfit(co2_concat['interpolated'],global_concat_1,1)
plt.plot(co2_concat['interpolated'],a*co2_concat['interpolated']+b, "r")
plt.title('Relationship between temperature anomalies and CO2 emission from 1958 to 2017')
plt.show()
```

```python
np.corrcoef(co2_concat['interpolated'],global_concat_1)
```
